```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.ar_model import AutoReg
import matplotlib.pyplot as plt
import yfinance as yf
import datetime as dt
```

```python
# Fetch stock data using yfinance
start = dt.datetime(2021, 6, 1)
end = dt.datetime(2022, 1, 5)
stk_data = yf.download('AXISBANK.NS', start=start, end=end)
```

```python
stk_data
```

```python
# Select relevant columns
stk_data = stk_data[["Open", "High", "Low", "Close"]]
```

```python
# Define column to be used
column = "Close"
```

```python
# Normalize the data
Ms = MinMaxScaler()
data1 = Ms.fit_transform(stk_data[[column]])
```

```python
# Split the data into training and testing sets
training_size = round(len(data1) * 0.98)
X_train = data1[:training_size]
X_test = data1[training_size:]
y_train = data1[:training_size]
y_test = data1[training_size:]
```

```python
print(f"X_train length: {X_train.shape}")
print(f"X_test length: {X_test.shape}")
print(f"y_train length: {y_train.shape}")
print(f"y_test length: {y_test.shape}")
```

```python
# Define the rmsemape function
def rmsemape(y_test, y_pred):
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    print(f"RMSE-Testset: {rmse}")
    print(f"maPe-Testset: {mape}")

# Define the graph function
def graph(actual, predicted, act_label, pred_label, title, x_label, y_label):
    plt.figure(figsize=(10, 5))
    plt.plot(actual, color='blue', label=act_label)
    plt.plot(predicted, color='green', label=pred_label)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.show()

i = 4
td = 'c'

# Fit the AutoReg model
model = AutoReg(X_train, lags=i, trend=td)
model_fit = model.fit()
# Make prediction
y_pred = model_fit.predict(len(X_train), len(X_train) + len(y_test) - 1)

# Calculate and print metrics
print(f"Lag={i}, Trend={td}")
rmsemape(y_test, y_pred)
print("************")
```

```python
# Plot the graph
graph(y_test, y_pred, "Actual", "Predicted", "AXISBANK Close Prices - AutoReg", "Days", "Prices")
```

```python
# Inverse transform the normalized data to get the original values
actual_stock_price_test_ori = Ms.inverse_transform(y_test)
predicted_stock_price_test_ori = Ms.inverse_transform(y_pred.reshape(-1, 1))
```

```python
# Plot the original scale graph
graph(actual_stock_price_test_ori, predicted_stock_price_test_ori, "Actual", "Predicted", "AXISBANK Close Prices - Original Scale", "Days", "Prices")
```

```python
# Calculate and print metrics for original scale
rmsemape(actual_stock_price_test_ori, predicted_stock_price_test_ori)
```

```python
# Forecast for the next step
forecast = model_fit.predict(len(data1), len(data1))
```

```python
# Inverse transform the forecast data
forecast_stock_price_test_ori = Ms.inverse_transform(forecast.reshape(-1, 1))
```

```python
# Plot the forecast
plt.figure(figsize=(10, 5))
plt.plot(actual_stock_price_test_ori, color='blue', label="Actual")
plt.plot(forecast_stock_price_test_ori, color='red', label="Predicted")
plt.title("AXISBANK Close Prices - Forecast")
plt.xlabel("Days")
plt.ylabel("Prices")
plt.legend()
plt.show()
```

```python
# Save forecast data to CSV
forecast_df = pd.DataFrame(forecast_stock_price_test_ori, columns=["Closefore"])
forecast_df.to_csv("CloseAR.csv", index=False)
```
