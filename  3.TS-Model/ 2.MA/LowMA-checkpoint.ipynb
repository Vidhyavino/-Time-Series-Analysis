```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import yfinance as yf
import datetime as dt
```

```python
# Fetch stock data using yfinance
start = dt.datetime(2021, 6, 1)
end = dt.datetime(2022, 1, 5)
stk_data = yf.download('AXISBANK.NS', start=start, end=end)
stk_data = stk_data[["Open", "High", "Low", "Close"]]
stk_data.to_csv("AXISBANK.NS.csv")
```

```python
stk_data
```

```python
# Preprocess the data
column = "Low"
Ms = MinMaxScaler()
data1 = Ms.fit_transform(stk_data[[column]])
print("Len:", data1.shape)
```

```python
# Split the data into training and test sets
training_size = round(len(data1) * 0.80)
print(training_size)
X_train = data1[:training_size]
X_test = data1[training_size:]
print("X_train length:", X_train.shape)
print("X_test length:", X_test.shape)
y_train = data1[:training_size]
y_test = data1[training_size:]
print("y_train length:", y_train.shape)
print("y_test length:", y_test.shape)
```

```python
# Suppress warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# Train and evaluate ARIMA models with different trends
trends = ['n', 't', 'c', 'ct']
orders = [(0, 0, 1), (0, 0, 2)]
```

```python
def rmsemape(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = mean_absolute_percentage_error(y_true, y_pred)
    print(f'RMSE: {rmse}, MAPE: {mape}')
    return rmse, mape
```

```python
def plot_graph(y_true, y_pred, title):
    plt.figure(figsize=(10, 5))
    plt.plot(y_true, label='Actual')
    plt.plot(y_pred, label='Predicted')
    plt.title(title)
    plt.xlabel('Days')
    plt.ylabel('Prices')
    plt.legend()
    plt.show()
```

```python
for td in trends:
    model = ARIMA(X_train, order=(0, 0, 2), trend=td)
    model_fit = model.fit()
    y_pred = model_fit.predict(len(X_train), len(data1) - 1)
    print(f"Trend={td}")
    rmsemape(y_test, y_pred)
    print("************")
```

```python
# Plotting actual vs predicted
plot_graph(y_test, y_pred, "AXISBANK.NS Low Price Prediction")
```

```python
# Inverse transform the scaled data
actual_stock_price_test_ori = Ms.inverse_transform(y_test)
predicted_stock_price_test_ori = Ms.inverse_transform(y_pred.reshape(-1, 1))
```

```python
# Plotting actual vs predicted original prices
plot_graph(actual_stock_price_test_ori, predicted_stock_price_test_ori, "AXISBANK.NS Low Price Prediction (Original)")
```

```python
# Forecast future prices
forecast = model_fit.predict(len(data1), len(data1) + 30)
forecast_stock_price_test_ori = Ms.inverse_transform(forecast.reshape(-1, 1))
```

```python
# Convert forecast to DataFrame and save to CSV
forecast_df = pd.DataFrame(forecast_stock_price_test_ori, columns=["Lowfore"])
forecast_df.to_csv("LowMA.csv", index=False)
```
